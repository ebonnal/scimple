```python
%matplotlib notebook
import scimple as scm
tab = scm.Table(scm.get_sample('xyz'),firstLine=2,lastLine=494) # 
moy = tab.reset_mapping().map(lambda _,line: (None,line[3]) if line[0] == 'C' and line[3] < 20 else None).reduce(lambda key,z : (key,sum(z)/len(z))).getMapping()[None][0]
to_plot = tab.reset_mapping().map(lambda i,line: (i,line[1:]+[moy]) if line[0] == 'C' and line[3] < 20 else None).getMappingAsTable(2)
scm.Plot(3).add(to_plot,1,2,3,markersize=4,plotType='x',coloredBy =lambda _,line:(line[3]-moy)**2).add(to_plot,1,2,4,plotType='.',markersize=3,coloredBy = "#880000")
scm.Plot(2).add(to_plot,1,2,markersize=4,plotType='x',coloredBy =lambda _,line:line[3],bg_color="#333355")
scm.show()
```
